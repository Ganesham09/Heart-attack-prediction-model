**MODELLING**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

x = df.iloc[:,:df.shape[1]-1].values
y = df['target'].values

x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7)
mod=LogisticRegression(max_iter=1000)
mod.fit(x_train,y_train)
```

```python
y_predicted = mod.predict(x_test)

print(y_test)
print(y_predicted)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
confusion_matrix(y_test, y_predicted)
```

```python
accuracy_score(y_test,y_predicted)
```

Building Predictive System

```python
input_data = (41,0,1,130,204,0,0,172,0,1.4,2,0,2)

## change the input data as numpy array
input_data_as_numpy_array = np.asarray(input_data)
## reshape the numpy array as we are predicting only one instance

input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = mod.predict(input_data_reshaped)

print(prediction)

if(prediction[0]==0):
    print('The person does not have heart disease')
else:
    print('The person have heart disease')
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_class = RandomForestClassifier()
rf_class.fit(x_train,y_train)

rf_class.score(x_train,y_train)
y_pred = rf_class.predict(x_test)
y_pred
```

```python
accuracy_score(y_test,y_pred)
```
